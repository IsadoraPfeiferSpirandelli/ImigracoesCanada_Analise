##1. Analisando as tendências de imigração do Brasil em um determinado período.

```python
#Criar uma variável para guardar todo o período/intervalo de tempo
anos = list(map(str, range(1980, 2014))) #range nao leva em conta o ultimo valor, entao pega 2013
anos
```

```python
#pegando somente a parte do Brasil criando uma nova variavel e usando o loc - essa propriedade consegue pegar rótulos específicos dentro de um dataframe
brasil = df.loc['Brasil', anos]
brasil
```

```python
#transformar a visualização acima em um data frame (dict = dictionary)

brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()} #criando colunas
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
#criando gráficos simples
import matplotlib.pyplot as plt
```

```python
#grafico de linhas - tempo

#confiurações do gráfico
plt.figure(figsize=(10, 5)) #tamanho da figura (largura, altura)
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes']) #colunas na ordem de eixo x e eixo y
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2015']) #para melhorar a visualização do eixo x vou agrupar em período de 5 anos

#rótulo dos eixos e título
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
plt.title('Imigração do Brasil para o Canadá')

plt.show() #sempre usar como ultima linha do meu código, pois o que vier depois nao será executado.
```

Percebe-se que a variável anos é independente e a imigrantes dependente

```python
# criar um gráfico de linhas comparando os números de imigrantes do Brasil e Argentina para o Canadá, que são os maiores países da América do Sul.
#preciso extrair os dados da Argentina também comparando com Brasil já
df_comparacao = df.loc[['Brasil', 'Argentina'], anos]
df_comparacao

#assim vou ter uma linha para cada país e as colunas contendo os números de imigrantes para cada ano, mas quero ao contrário, então:
df_comparacao = df_comparacao.T
df_comparacao.head()
#agora tenho uma coluna com os anos e duas colunas para cada país contendo o número de imigrantes.
```

```python
#plotando o gráfico de linhas da comparação

plt.plot(df_comparacao['Brasil'],label = 'Brasil')
plt.plot(df_comparacao['Argentina'],label ='Argentina')


plt.title('Imigração do Brasil e Argentina para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')

plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2015'])
plt.legend() #vai me dar uma legenda

plt.show()
```

##Conclusão do gráfico 2
A partir do gráfico acima, nota-se que no início as taxas de imigração no Brasil e na Argentina eram bastante semelhantes. Mas, a partir dos anos 2000 nós podemos observar a crescente diferença entre eles.
De 2000 à 2005 os números de imigrações na Argentina atingiu seu máximo neste período e começa a decair por volta de 2005.
Já no Brasil, observa-se que no período de 2000 à 2005 ele vem crescendo gradativamente e aos poucos, até que a partir de 2005 à 2010 ele atinge seu máximo nas imigrações e aí começa a decair novamente.

```python
#criação de figuras
#fig: permite que uma figura seja criada (um espaço em branco)
#ax: represenda os eixos
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # eixo x com periodo de 5 em 5

ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Número de Imigrantes')

plt.show()
```

```python
#criar dois graficos ao mesmo tempo em uma figura

fig, axs = plt.subplots(1, 2, figsize=(15, 5)) # onde 1 = linhas e 2 = quantidade de subplots

#criando um de linhas
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5)) #frequencia dos anos
axs[0].grid() #criar uma grade no gráfico para melhor entendimento das informações

#criando um boxplot
axs[1].boxplot(dados_brasil['imigrantes']) # no boxplot coloco só os dados numéricos
axs[1].set_title('Boxplot da Imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de Imigrantes')
axs[1].grid()

plt.show()
```

##Conclusão da figura acima
O gráfico de linhas me mostra as tendências de imigração do Brasil para o Canadá ao longo dos anos e o boxplot me mostra algumas estatísticas dessa imigração.
Boxplot me dá valores de mínimo e máximo, as estatísticas em 25%, 50$ e 75% do número de imigrantes.

#2. Analisando tendências de imigração dos países da América do Sul

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10)) #2 subplots por linha num total de duas linhas
fig.subplots_adjust(hspace=0.5, wspace=0.3) #espaçamento entre os subplots
fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá\n1980 a 2013')
#vou usar meu dataframe original (df) para pegar as infos dos outros países

#Gráfico Brasil
axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

#Gráfico Colômbia
axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

#Gráfico Argentina
axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

#Gráfico Peru
axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')


#colocando todos os graficos com frenquencia 5 nos anos
for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.grid()

#colocando os labels para todos
for ax in axs.flat:
  ax.set_xlabel('Ano')
  ax.set_ylabel('Número de Imigrantes')

#deixar os valores minimos e maximos padronizados para que o entendimento do gráfico não seja distorcido
ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

A partir da figura acima, entre os países apresentados, é possível perceber que a Colômbia é o país da América do Sul com maior número de imigração em disparada para o Canadá.

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
df_lojas = pd.DataFrame(vendas_2022, index=lojas)
df_lojas
```

```python
# Criar a figura e os subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

# Ajustar os espaçamentos entre os subplots
plt.subplots_adjust(wspace=0.3, hspace=0.4)

# Adicionando um título geral para os subplots
fig.suptitle('Vendas no período de janeiro a dezembro de 2022 nas lojas A,B,C e D')

# Adicionar os gráficos em cada um dos subplots
axs[0, 0].plot(df_lojas.loc['A'])
axs[0, 0].set_title('Vendas na loja A')

axs[0, 1].plot(df_lojas.loc['B'])
axs[0, 1].set_title('Vendas na loja B')

axs[1, 0].plot(df_lojas.loc['C'])
axs[1, 0].set_title('Vendas na loja C')

axs[1, 1].plot(df_lojas.loc['D'])
axs[1, 1].set_title('Vendas na loja D')

# Adicionando rótulos para os eixos X e Y
for ax in axs.flat:
    ax.set_xlabel('Mês')
    ax.set_ylabel('Número de vendas')

# Exibir a figura
plt.show()
```

```python
#criando função do grafico
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h', palette=palette)
  ax.set_title('Top 10 países com mais imigrantes para o Canadá\n1980 a 2013', loc='left', fontsize=16)
  ax.set_xlabel('Número de Imigrantes', fontsize=14)
  ax.set_ylabel('')
  plt.show()
```

```python
#fazendo alterações
fig = px.line(dados_brasil, x='ano', y='imigrantes',
              title='Imigração do Brasil para o Canadá\nde 1980 a 2013')

fig.update_traces(line_color='green', line_width=3)

fig.update_layout(xaxis_title='Ano',
                  yaxis_title='Número de Imigrantes',
                  width=1000, height=500, #alterando largura e altura
                  xaxis={'tickangle':-45}, #alterando a rotação dos numeros no eixo x
                  font_family='Arial',
                  font_size=14,
                  font_color='grey',
                  title_font_color='black',
                  title_font_size=16)
fig.show()
```

```python
#criando grafico interativo para america do sul
df_america_sul_clean = america_sul.drop(['Continente', 'Região', 'Total'], axis=1)
#fazendo uma transposição
america_sul_final = df_america_sul_clean.T
america_sul_final
```

```python
america_sul_final.head()
```

```python
fig = px.line(america_sul_final, x=america_sul_final.index, y=america_sul_final.columns,
              color='País',
              title='Imigração dos países da América do Sul para o Canadá\nde 1980 a 2013')

fig.update_layout(xaxis_title='Ano',
                  yaxis_title='Número de Imigrantes',
                  xaxis={'tickangle':-45}, #alterando a rotação dos numeros no eixo x
                  width=1000, height=500, #alterando largura e altura
                  font_family='Arial',
                  font_size=14,
                  font_color='grey',
                  title_font_color='black',
                  title_font_size=16)
fig.show()
```
